##Classify Movies Based on Ratings

```python
def classify_rating(rating):
  if rating>=8.0:
    return "Excellent"
  elif rating >=7.0:
    return "Good"
  elif rating >= 6.0:
    return "Average"
  else:
    return "Poor"
```

```python
df['rating_class'] = df['Rating'].apply(classify_rating)
df
```
